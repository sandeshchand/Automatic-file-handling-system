```python
import os
import shutil
import docx
import logging as lg

lg.basicConfig(filename="file_handling.log",level=lg.INFO, format='%(asctime)s,%(message)s')
def file_Handling(file_path):
    try:
        os.chdir(file_path)
        for file_name in os.listdir():
            if('docx' in file_name)and (os.path.isfile(file_name)==True):
                lg.info('title',file_name)
                document=docx.Document(file_name)
        #print(document)
                full_data=''
                fullText=[]
                for para in document.paragraphs:
                    fullText.append(para.text)
                    full_data='\n'.join(fullText)
            #print(full_data)
 
                for iter in file_name:
             #print(iter,end= " ")
        
            
                    if iter.isdigit()== True:
                        folder_name=file_name.split(iter)[0]
                                      
                        path= file_path +"\\"+folder_name
                        if os.path.isdir(path)==False:
                            os.mkdir(path)
                        os.chdir(path)
                    
                        try:
                            my_doc=docx.Document('Final'+folder_name+'.docx')
                        except:
                            my_doc=docx.Document()
                            my_doc.save('Final'+ folder_name+'.docx')
            
                        my_doc.add_paragraph('\n\n filename is '+ file_name)
                        my_doc.add_paragraph(full_data)
                        my_doc.save('Final'+folder_name+ '.docx')
        
                        os.chdir(file_path)
                        shutil.move(file_name,path)
                        lg.info('File move to folder  done!!')
            
                        break
                    else:
                        pass

    
    except Exception as e:
        print("Required file {} doesnot have digit to split".format(file_name))
        lg.exception((str(e)))    
def file_Path():
    file_path=input('Please enter the current path of file.')
    file_Handling(file_path)

if __name__=="__main__":
    file_Path()
               
```
